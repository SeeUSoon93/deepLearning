### 머신러닝과 딥러닝의 차이 확인하기
- 동일한 데이터를 가지고 머신러닝, 딥러닝 학습 후 예측 진행
- 공부시간에 따른 학생의 수학성적을 예측하는 모델링

### 머신러닝모델, 딥러닝 모델의 차이점
- 머신러닝 sklearn
  - 모델객체생성(완성된 객체를 사용) -> 모델학습 -> 모델예측 -> 모델평가
  - 완제품 로봇 : 팔정도만 움직일 수 있음(하이퍼파라미터 조절)
- 딥러닝 tensorflow
  - 모델객체생성(모델을 직접 구성) -> 모델학습 -> 모델예측 -> 모델평가
  - 레고블럭 : 사용자가 직접 설계
  - 다양한 결과를 만들어낼 수 있음

### 문제와 답으로 분리
- 입력특성:1개 (studytime)
- 정답데이터:G3

```python
#문제데이터 X, 정답데이터 y 대입
X = data['studytime'] # 딥러닝까지 진행할 데이터 -> 일단은 1차원
y = data['G3']
```

```python
# 크기확인 (shape)
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
# train, test 분리
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 10)
```

```python
# 선형모델 불러와서 학습 및 예측
from sklearn.linear_model import LinearRegression
# 회귀모델의 평가지표 도구 (mse)
from sklearn.metrics import mean_squared_error
```

```python
#1. 모델생성
lr_model = LinearRegression()
#2. 모델학습
lr_model.fit(pd.DataFrame(X_train), y_train)
#3. 모델예측
pre = lr_model.predict(pd.DataFrame(X_test))
#4. 모델평가
mse = mean_squared_error(y_test, pre)
print(f"Mean Squared Error (MSE): {mse}")
```
